```python
import pandas as pd, numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
SAT = pd.read_csv('SAT_Results.csv')
```

```python
SAT = SAT[SAT.iloc[:,2] != 's']

SAT['Total'] = Reading + Writing + Math
SAT['Above'] = SAT['Total'] > 1490

SAT_a = SAT.query('Above == True')
SAT_b = SAT.query('Above == False')

def num(s):
    try:
        return int(s)
    except ValueError:
        return float(s)


Size_a = SAT_a.iloc[:,2].map(num)
Reading_a = SAT_a.iloc[:,3].map(num)
Math_a = SAT_a.iloc[:,4].map(num)
Writing_a = SAT_a.iloc[:,5].map(num)

Size_b = SAT_b.iloc[:,2].map(num)
Reading_b = SAT_b.iloc[:,3].map(num)
Math_b = SAT_b.iloc[:,4].map(num)
Writing_b = SAT_b.iloc[:,5].map(num)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(10, 8))
ax = Axes3D(fig)
ax.scatter(Reading_a, Math_a, Writing_a, s = Size_a, c = 'g')
ax.scatter(Reading_b, Math_b, Writing_b, s = Size_b, c = 'r')
ax.set_title("Average SAT Scores in NYC High Schools", fontsize = 20)
ax.set_xlabel("Avg. Reading Score", fontsize = 18)
ax.set_ylabel("Avg. Math Score", fontsize = 18)
ax.set_zlabel("Avg. Writing Score", fontsize = 18)
plt.show()
```

## The above 3D plot demonstrates the average scores each NYC public school's student achieve in the SAT exam in 2015, based off NYC Open Data. The shape of each point represents the number of students taken SAT exams in each school. 

## According to http://blog.prepscholar.com/what-is-the-average-sat-score , the average SAT score in 2015 was 1490. Using 1490 as a benchmark, the green dots are schools with the sum of the average scores for each part of the SAT exam above the national average, and the red dots are the schools that do not reach the number. 


## The plot also shows that, three average scores have strong correlations, meaning that one school's average scores for Reading, Mathematics, and Writing sections are probably close to each other.
